```python
# --- CODE 1: Preprocessing and saving dataset ---
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Load original dataset
df = pd.read_csv('smart_grid_stability_augmented.csv')

# Separate features and targets
X = df.drop(columns=['stab', 'stabf'])
y_continuous = df['stab']
y_discrete = df['stabf'].map({'stable': 1, 'unstable': 0})

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Combine preprocessed features and targets
preprocessed_df = pd.DataFrame(X_scaled, columns=X.columns)
preprocessed_df['stab'] = y_continuous
preprocessed_df['stabf'] = y_discrete

# Save to new Excel file
preprocessed_df.to_csv('preprocessed_smart_grid_data.csv', index=False)

print("Preprocessing completed and saved to 'preprocessed_smart_grid_data.csv'")
```
